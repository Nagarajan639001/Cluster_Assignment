```python
#Import library
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# Selct the necessary columns
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
#apply the cluster No,create the Model & predict (i.e. grade)
```

```python
from sklearn.cluster import DBSCAN
DB=DBSCAN(eps=2,min_samples=2).fit(X)
Y_DB=DB.fit_predict(X)
```

```python
Y_DB
```

```python
#create the empty dataset
```

```python
supervised=dataset
```

```python
supervised
```

```python
# added the group coloums in empty dataset
```

```python
supervised['Cluster_group']=Y_DB
```

```python
supervised
```

```python
#Save the New create file .csv
```

```python
supervised.to_csv("DBSCAN.csv",index =False)
```
