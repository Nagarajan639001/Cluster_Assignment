```python
#Import library
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# Selct the necessary columns
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
# Identify the number of cluster
```

```python
import scipy.cluster.hierarchy as sch
dendrogram= sch.dendrogram(sch.linkage(X,method='ward'))

plt.title("Dendragram")
plt.xlabel("Customer")
plt.ylabel("Education distances")
plt.show
```

```python
#apply the cluster No,create the Model & predict (i.e. grade)
```

```python
from sklearn.cluster import AgglomerativeClustering
clusmodle = AgglomerativeClustering(n_clusters=5)
label = clusmodle .fit_predict(X)
```

```python
label
```

```python
#create the empty dataset
```

```python
supervised=dataset
```

```python
supervised
```

```python
# added the group coloums in empty dataset
```

```python
supervised['Cluster_group']=label
```

```python
supervised
```

```python
#Save the New create file .csv
```

```python
supervised.to_csv("Agglomerative.csv",index =False)
```
