```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
# Selct the necessary columns
```

```python
X=dataset.iloc[:,[3,4]].values
```

```python
X
```

```python
#apply the cluster No,create the Model & predict (i.e. grade)
```

```python
from sklearn.cluster import AffinityPropagation
aff=AffinityPropagation(damping=0.75,max_iter=200,convergence_iter=15,copy= True,affinity='euclidean',preference= None,
                        verbose = False)
Y_aff=aff.fit(X)
```

```python
#create the empty dataset
```

```python
supervised=dataset
```

```python
supervised
```

```python
# added the group coloums in empty dataset
```

```python
supervised['Cluster_group']=Y_aff
```

```python
supervised
```

```python
#Save the New create file .csv
```

```python
supervised.to_csv("Affinit Propagation.csv",index =False)
```
